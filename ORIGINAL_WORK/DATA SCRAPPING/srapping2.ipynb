**DEEL**

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Configuration des options du navigateur Chrome
chrome_options = Options()
chrome_options.add_argument("--headless")  # Exécution sans fenêtre
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Instanciation du pilote Chrome avec les options configurées
driver = webdriver.Chrome(options=chrome_options)

# URL de la page de critiques
base_url = 'https://www.capterra.fr/reviews/205015/deel'

# Liste pour stocker les données extraites
donnees = []

# Nombre total de pages de critiques
nombre_pages = 21

# Boucle pour parcourir toutes les pages de critiques
for page in range(1, nombre_pages + 1):
    # Construire l'URL de la page de critique actuelle
    url_page = f"{base_url}?page={page}"

    # Aller à l'URL de la page de critique
    driver.get(url_page)

    WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, 'container'))
)



    # Trouver les éléments de commentaire par XPath
    # Trouver les éléments de commentaire par classe
    reviews = driver.find_elements(By.XPATH, "//h3[@class='h5 fw-bold']")
    # Trouver les éléments de note par classe
    ratings = driver.find_elements(By.XPATH, "//span[@class='ms-1']")

    # Parcourir les éléments et extraire les commentaires et les notes
    for review, rating in zip(reviews, ratings):
        commentaire = review.text
        note = rating.text
        # Ajouter les données extraites à la liste
        donnees.append({'Plateforme': 'Deel', 'Avis': commentaire, 'Notes sur 5': note})

# Créer un DataFrame à partir de la liste de dictionnaires
df = pd.DataFrame(donnees)

# Afficher le DataFrame
print(df)

# Voir le shape
print("Shape du DataFrame:", df.shape)

# Convertir en fichier csv
df.to_csv('deel.csv', index=False)

# Fermer le navigateur
driver.quit()
```
