#Scrapping des données sur Trustpilot

*Scrapping de Deel.com*

```python
from time import sleep
import requests
from google.colab import files
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Définir la fonction de récupération des données
def soup2list(src, list_, attr=None): # Prendre en entrée la source et la liste dans laquelle les données externes seront stockés
  if attr: # Si l'attribut à été spécifié on l'extrait
    for val in src:
      list_.append(val[attr])
  else: # Sinon on extrait le texte
    for val in src:
      list_.append(val.get_text())
```

```python
# Définir les features à extraires
users = []
userReviewNum = []
ratings = []
locations = []
dates = []
reviews = []

# Les pages
from_page = 1
to_page = 187

# Plateformes
company = 'www.deel.com'
```

```python
# Fonction de scrapping
for i in range(from_page, to_page+1):
  result = requests.get(fr"https://www.trustpilot.com/review/{company}?page={i}")
  soup = BeautifulSoup(result.content)

   # Scrapper les données
  soup2list(soup.find_all('span', {'class','typography_heading-xxs__QKBS8 typography_appearance-default__AAY17'}), users)
  soup2list(soup.find_all('div', {'class','typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l styles_detailsIcon__Fo_ua'}), locations)
  soup2list(soup.find_all('span', {'class','typography_body-m__xgxZ_ typography_appearance-subtle__8_H2l'}), userReviewNum)
  soup2list(soup.find_all('div', {'class','styles_reviewHeader__iU9Px'}), dates)
  soup2list(soup.find_all('div', {'class','styles_reviewHeader__iU9Px'}), ratings, attr='data-service-review-rating')
  soup2list(soup.find_all('div', {'class','styles_reviewContent__0Q2Tg'}), reviews)

  # Temps de pause
  sleep(1)

```

```python
# Transformer nos données en dataframe
truspilot_deel_data = pd.DataFrame(
    {
        'Plateforme': 'deel.com',
        'Utilisateur':users,
        'Nombre de commentaires': userReviewNum,
        'Position Géographique': locations,
        'Date':dates,
        'Avis': reviews,
        'Notes sur 5': ratings
    }
)
```

```python
# Voir le datasets
truspilot_deel_data
```

```python
# Convertir en fichier csv
truspilot_deel_data.to_csv('deel.csv', index=False)
files.download('deel.csv')
```
